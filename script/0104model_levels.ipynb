```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read in the data
df = pd.read_csv('/Users/changyichieh/Documents/Hot Jupiter/Hot_Jupiter_Github/hot_jupiter/Parmentier 2/PTprofile(20bar)(Hd209458b)(0103).csv')
df.columns
```

```python
# Read in the data
new_df = pd.DataFrame()

new_df['N'] = df['N']
new_df['P'] = df['P(Pa)']


new_df['ak'] = df['P(Pa)']
new_df['bk'] = 0
new_df['ph'] = df['P(Pa)']/100
# pf is the midpoint pressure of each layer
new_df['pf'] = (new_df['P'] + new_df['P'].shift(1)) / 2
new_df.loc[0, 'pf'] = new_df.loc[0, 'P']/2  # or any other method you find suitable

# insert a N=0 row
new_df.loc[-1] = 0  # adding a row
new_df.index = new_df.index + 1  # shifting index
new_df = new_df.sort_index()  # sorting by index

new_df['pf']/=100

del new_df['P']
# del new_df['N']

new_df
```

```python
# save the new_df to a space-separated text file
np.savetxt('/Users/changyichieh/Documents/Hot Jupiter/Hot_Jupiter_Github/hot_jupiter/Parmentier 2/model_levels_Hd209458b.txt', new_df,
            fmt=['%d', '%10.2f', '%10.2f', '%10.2f', '%10.2f'], delimiter='\t', header='N ak bk ph pf', comments='')
```
