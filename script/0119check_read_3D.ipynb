```python
import numpy as np
import pandas as pd

def read_3d_array_from_file(filename, mp, nx, ny):
    try:
        # Open the file and read the lines
        with open(filename, 'r') as file:
            lines = file.readlines()

        # Initialize the 3D array
        zzz = np.zeros((mp, nx, ny))

        # Skip the header line
        line_index = 1

        for ip in range(mp):
            # Skip the slice header
            line_index += 1

            for ix in range(nx):
                # Read the data line and convert to float
                data_line = lines[line_index].strip().split()
                zzz[ip, ix, :] = [float(value) for value in data_line]

                line_index += 1

        return zzz

    except IOError as e:
        print(f'Error opening or reading file: {e}')
        return None
    except Exception as e:
        print(f'Error processing file: {e}')
        return None

# Usage
filename = '/Users/changyichieh/Documents/Hot Jupiter/epsilon_prime_3D_profile.txt'
mp, nx, ny = 60, 96, 48  # Replace with actual dimensions
zzz = read_3d_array_from_file(filename, mp, nx, ny)

if zzz is not None:
    # zzz now contains the 3D array
    print(zzz)
```
